```python
def generate_table_creation_statements(json_data, parent_key=None):
    statements = []

    if isinstance(json_data, dict):
        for key, value in json_data.items():
            new_parent_key = f"{parent_key}_{key}" if parent_key else key
            statements.extend(generate_table_creation_statements(value, new_parent_key))

    elif isinstance(json_data, list):
        for item in json_data:
            statements.extend(generate_table_creation_statements(item, parent_key))

    elif isinstance(json_data, int):
        statements.append(f"{parent_key} INT,")

    elif isinstance(json_data, str):
        statements.append(f"{parent_key} VARCHAR(255),")

    elif isinstance(json_data, float):
        statements.append(f"{parent_key} FLOAT,")

    elif isinstance(json_data, bool):
        statements.append(f"{parent_key} BOOLEAN,")

    elif json_data is None:
        statements.append(f"{parent_key} NULL,")

    return statements

# Example JSON data with dynamic keys
example_json_data = {
    "college": {
        "name": "Sample College",
        "location": "Cityville",
        "courses": [
            {
                "course_id": 101,
                "course_name": "Introduction to Computer Science",
                "professor": {
                    "professor_id": 1,
                    "professor_name": "Dr. Smith",
                    "department": "Computer Science"
                },
                "students": [
                    {
                        "student_id": 1001,
                        "student_name": "John Doe",
                        "major": "Computer Science",
                        "enrollment_status": "Enrolled",
                        "grades": {
                            "midterm": 85,
                            "final": 92
                        }
                    }
                ]
            }
        ],
        "additional_info": {
            "start_date": "2022-09-01",
            "campus_size": 500,
            "has_sports_facility": True,
            "campus_location": None
        }
    }
}

# Generate table creation statements
table_creation_statements = generate_table_creation_statements(example_json_data)

# Print the generated SQL statements
for statement in table_creation_statements:
    print(statement)
```
